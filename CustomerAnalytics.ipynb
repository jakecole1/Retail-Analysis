### Data Engineering
The dataset used for this project is the "Online Retail 2" dataset from the UCI machine learning repository.

The first step is to initialize a SparkSession and read in the retail dataset using the `spark.read.csv()` method. This method reads in a CSV file and returns a DataFrame, which is a distributed collection of data organized into named columns.

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import to_date,col

# Initialize Spark Session
spark = SparkSession.builder \
    .appName("Customer Behavior Analytics") \
    .getOrCreate()



retail_df = spark.read.csv('/Users/jacobcole/Downloads/online_retail_II.csv', header=True, inferSchema=True)
retail_df = retail_df.na.drop(subset=["Invoice", "StockCode", "Description", "Quantity", "InvoiceDate", "Price", "Customer ID", "Country"])


retail_df = retail_df.withColumn('InvoiceDate', to_date(col('InvoiceDate'), 'MM/d/yyyy H:mm'))


retail_df = retail_df.withColumn('TotalPrice', col('Quantity') * col('Price'))


retail_df = retail_df.select("Invoice", "StockCode", "Description", "Quantity", "InvoiceDate", "Price", "Customer ID", "Country", "TotalPrice")


retail_df.show(5)
```

### Cleaning data and feature engineering


```python
from pyspark.sql.window import Window

monetary_df = retail_df.groupBy('Customer ID').sum('TotalPrice').withColumnRenamed('sum(TotalPrice)', 'Monetary')

frequency_df = retail_df.groupBy('Customer ID', 'Invoice').count().groupBy('Customer ID').count().withColumnRenamed('count', 'Frequency')

```

```python
from pyspark.sql.functions import max,datediff, lit,hour,dayofweek
spark.conf.set("spark.sql.legacy.timeParserPolicy", "LEGACY")

max_date_df = retail_df.agg(max('InvoiceDate').alias('MaxInvoiceDate'))

max_date = max_date_df.first()['MaxInvoiceDate']
recency_df = retail_df.groupBy('Customer ID').agg(max('InvoiceDate').alias('LastPurchase'))
recency_df = recency_df.withColumn('Recency', datediff(lit(max_date), 'LastPurchase'))

rfm_df = monetary_df.join(frequency_df, 'Customer ID').join(recency_df, 'Customer ID')
rfm_df.show(5)
```

```python
retail_df = retail_df.withColumn('InvoiceHour', hour(col('InvoiceDate')))
retail_df = retail_df.withColumn('InvoiceDayOfWeek', dayofweek(col('InvoiceDate')))
```

## Customer Data Transformation and Clustering Visualization

This section of the code performs several key data transformation and analysis tasks on the customer dataset:

1. **Feature Assembly and Scaling**: We combine the 'Recency', 'Frequency', and 'Monetary' columns into a single feature vector and scale the data to normalize the feature space.
2. **Dimensionality Reduction with PCA**: Principal Component Analysis is applied to the scaled data to reduce its dimensions to two principal components, making it suitable for visualization.
3. **K-Means Clustering**: The K-Means algorithm is employed to segment customers into four distinct groups based on the scaled features.
4. **Prediction and Plotting**: We predict the cluster group for each customer and plot the results in a scatter plot, using the two principal components for the axes, to visualize the customer segmentation.


```python
from pyspark.ml.feature import VectorAssembler, StandardScaler
from pyspark.ml.clustering import KMeans
from pyspark.ml.feature import PCA
import pandas as pd

# Assemble features and scale them
vec_assembler = VectorAssembler(inputCols=['Recency', 'Frequency', 'Monetary'], outputCol='features')
rfm_features_df = vec_assembler.transform(rfm_df)
scaler = StandardScaler(inputCol="features", outputCol="scaledFeatures")
rfm_scaled_df = scaler.fit(rfm_features_df).transform(rfm_features_df)

# Apply PCA to reduce dimensions
pca = PCA(k=2, inputCol="scaledFeatures", outputCol="pcaFeatures")
pca_model = pca.fit(rfm_scaled_df)
pca_result = pca_model.transform(rfm_scaled_df)

# Fit the K-means model
kmeans = KMeans(featuresCol='scaledFeatures', k=4)
kmeans_model = kmeans.fit(rfm_scaled_df)

# Make predictions and add them to the pca_result DataFrame
predictions = kmeans_model.transform(pca_result)

predictions_df = predictions.select('pcaFeatures', 'prediction')



pandas_df = predictions.toPandas()
pandas_df['pcaFeatures'] = pandas_df['pcaFeatures'].apply(lambda x: x.toArray().tolist())

# Extract the PCA features for plotting
pandas_df[['PC1', 'PC2']] = pd.DataFrame(pandas_df['pcaFeatures'].tolist(), index=pandas_df.index)


plt.figure(figsize=(10, 8))
sns.scatterplot(x="PC1", y="PC2", hue="prediction", data=pandas_df, palette="viridis")
plt.title('Customer Segmentation')
plt.show()
```

## Customer Churn Analysis

In the following code, we focus on the analysis of customer churn within the retail dataset:

1. **Determine the Latest Invoice Date**: We find the most recent date in the dataset to use as a reference point for calculating recency.
2. **Calculate Recency for Each Customer**: We group the data by customer ID and calculate the number of days since their last purchase, using the latest invoice date as a benchmark.
3. **Label Customers as Churned or Active**: We introduce a churn label, marking customers as churned (`1`) if they have not made a purchase in the last 60 days, otherwise they are marked as active (`0`).
4. **Join Churn Data to Original DataFrame**: Finally, we join this churn information back to the original retail dataset to enrich it with churn status and recency information for each customer.


```python
from pyspark.sql.functions import when
max_date_df = retail_df.agg(max('InvoiceDate').alias('MaxInvoiceDate'))
max_date = max_date_df.first()['MaxInvoiceDate']


churn_period = 60


recency_df = retail_df.groupBy('Customer ID').agg(max('InvoiceDate').alias('LastPurchase'))
recency_df = recency_df.withColumn('Recency', datediff(lit(max_date), 'LastPurchase'))

# Add a churn label: 1 for churned customers, 0 for active customers
recency_df = recency_df.withColumn('Churn', when(col('Recency') > churn_period, 1).otherwise(0))

retail_df = retail_df.join(recency_df.select('Customer ID', 'Churn','Recency'), on='Customer ID', how='left_outer')
```

```python
churn_pd = retail_df.select('Churn').toPandas()
sns.barplot(x="Churn", y="count", data=churn_pd.value_counts().reset_index().rename(columns={0: 'count'}))

plt.show()
```

## Random Forest Classification for Churn Prediction

This section of the code prepares and applies a Random Forest Classifier to predict customer churn:

1. **Data Joining**: We merge `retail_df` with `frequency_df` and `monetary_df` to collate all relevant customer features into a single DataFrame, keyed by 'Customer ID'.
2. **Feature Assembly**: A `VectorAssembler` is used to combine selected columns ('Quantity', 'Price', 'TotalPrice', 'InvoiceDayOfWeek') into a feature vector required for machine learning models.
3. **Data Splitting**: The prepared dataset is randomly split into a training set (70%) and a testing set (30%).
4. **Model Training**: A Random Forest model is trained on the training data to predict the 'Churn' label.
5. **Model Evaluation**: The trained model is then used to make predictions on the test data, and the first five predictions are displayed to provide a quick evaluation.


```python
from pyspark.ml.classification import RandomForestClassifier
from pyspark.ml.feature import VectorAssembler
from pyspark.ml import Pipeline

# Assuming retail_df, frequency_df, and monetary_df are already defined and preprocessed
joined_df_1 = retail_df.join(frequency_df.select('Customer ID', 'Frequency'), on='Customer ID', how='left')
joined_df = joined_df_1.join(monetary_df.select('Customer ID', 'Monetary'), on='Customer ID', how='left')

# Assemble features
feature_cols = ['Quantity','Price','TotalPrice','InvoiceDayOfWeek']
assembler = VectorAssembler(inputCols=feature_cols, outputCol="features")
churn_df = assembler.transform(joined_df)


(train_data, test_data) = churn_df.randomSplit([0.7, 0.3])

# Build the Random Forest model
rf = RandomForestClassifier(labelCol="Churn", featuresCol="features", numTrees=20) 
rfModel = rf.fit(train_data)

# Evaluate the model
predictions = rfModel.transform(test_data)
predictions.select('Customer ID', 'prediction').show(5)
```

### Model Evaluation:


```python
from pyspark.ml.evaluation import MulticlassClassificationEvaluator

# Let's assume 'predictions' is the DataFrame that contains the output from the model
evaluator = MulticlassClassificationEvaluator(
    labelCol="Churn", predictionCol="prediction", metricName="accuracy")

# Compute the accuracy on the test data
accuracy = evaluator.evaluate(predictions)

print("Test Accuracy = %g" % accuracy)

# Now plot the accuracy using matplotlib
plt.bar(['Random Forest'], [accuracy])
plt.xlabel('Model')
plt.ylabel('Accuracy')
plt.title('Model Accuracy')
plt.ylim(0, 1)  # Assuming accuracy is between 0 and 1
plt.show()
```

```python
trainingSummary = rfModel.summary
# Precision-Recall Curve Data
pr = trainingSummary.pr.toPandas()
plt.plot(pr['recall'], pr['precision'])
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.show()
```
